## Car Price Suggestion

Problem Statement:

A Chinese automobile company Geely Auto aspires to enter the US market by setting up their manufacturing unit in the United States.

They want to understand  the factors on which the pricing of cars depends. Specifically, they want to understand the factors affecting the pricing of cars in the American market by analysing car data of the existing players in the US market. 


Essentially, the company wants —


- To understand which variables are significant in predicting the price of a car

- How well those variables describe the price of a car.

Essentially, the company wants —


1. Importing and understanding the data.

2. Data cleaning and preparation

3. Validation of data for Linear Regression applicability.

    1.	Frequency distribution of Dependent variable for outlier/spikes 
    2.	Multi co-linearity test of dependent variable presence using VIF method
    3.	Understanding insignificant dependent variable presence using p value test
    4.	Understand presence of Auto co-linearity using Durbin Watson ( D) value
    5.	Understand Heteroskedasticity ??

4. Dividing the selected dataset into train and test with 70:30 split

5.	Selection of  useful variable by applying RFE(Recursive Feature Elimination ) method

6.	Iterative linear regression model building

7.	Model Validation and result interpretation

### Data Cleaning and Preparation

```python
# Converting two level categorical veriable to 1 and 0
carprice_multi1=carprice_multi
carprice_multi1['fueltype'] = carprice_multi1['fueltype'].map({'gas': 1, 'diesel': 0})
carprice_multi1['aspiration'] = carprice_multi1['aspiration'].map({'std': 1, 'turbo': 0})
carprice_multi1['doornumber'] = carprice_multi1['doornumber'].map({'two': 1, 'four': 0})
carprice_multi1['enginelocation'] = carprice_multi1['enginelocation'].map({'front': 1, 'rear': 0})
carprice_multi1.head()
```

```python
#### Validation for Multi co-linearity test of dependent variable presence using VIF method
```

```python
import statsmodels.api as sm  
# UDF for calculating vif value
def vif_cal(input_data, dependent_col):
    vif_df = pd.DataFrame( columns = ['Var', 'Vif'])
    x_vars=input_data.drop([dependent_col], axis=1)
    xvar_names=x_vars.columns
    for i in range(0,xvar_names.shape[0]):
        y=x_vars[xvar_names[i]] 
        x=x_vars[xvar_names.drop(xvar_names[i])]
        rsq=sm.OLS(y,x).fit().rsquared  
        vif=round(1/(1-rsq),2)
        vif_df.loc[i] = [xvar_names[i], vif]
    return vif_df.sort_values(by = 'Vif', axis=0, ascending=False, inplace=False)

vif_cal(input_data=carprice_multi3, dependent_col="FinalPrice")
# Many independent variables have high inter dependency that need to be removed.
```

It is extremely important to rescale the variables so that they have a comparable scale. 
There are twocoon ways of rescaling 
1. Normalisation (min-max scaling) and 
2. standardisation (mean-o, sigma-1) 
Let's try normalisation

```python
# Normalization is not requried.
#defining a normalisation function 
#def normalize (x): 
#    return ( (x-np.min(x))/ (max(x) - min(x)))
                                            
#defining a standardization function 
def standardize (x): 
    return ( (x-np.mean(x))/ np.std(x))

#applying normalize ( ) to all columns 
#carprice_multi3 = carprice_multi3.apply(normalize)
carprice_multi3 = carprice_multi3.apply(standardize)
#carprice_multi3.head()
```
