```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display, HTML
from matplotlib.gridspec import GridSpec

#loading the companies csv file into rounds2 data frame
companies = pd.read_csv("Companies.csv",encoding = "ISO-8859-1")

companies = companies.applymap(lambda s:s.lower() if type(s) == str else s) #convert the entire data-frame to lower string
companies['permalink'].replace('/organization/','',regex=True,inplace=True)
# companies['permalink'] = companies['permalink'].str.replace('\W', '') #removing all special characters.
companies = companies[companies['status'] != 'closed'] # removing companies that are not operating anymore.
companies = companies.dropna(subset=['category_list']) # Dropping rows where category is absent.
# companies['unique_key'] = companies['region'].astype(str)+companies['permalink'].astype(str) + companies['name'].astype(str) + companies['founded_at'].astype(str)
display(companies[companies.duplicated(subset='permalink',keep = False)].shape[0])
display(companies)

display(companies.nunique()) #counting the number of unique companies in the sheet
```

```python
#loading the rounds2 csv file into data frame
rounds2 = pd.read_csv("rounds2.csv",encoding = "ISO-8859-1")

rounds2 = rounds2.applymap(lambda s:s.lower() if type(s) == str else s) #convert the entire data-frame to lower string
rounds2['company_permalink'].replace('/organization/','',regex=True,inplace=True) #removing unnecessary strings from the columns

rounds2['funding_round_permalink'].replace('/funding-round/','',regex=True,inplace=True)
# rounds2['company_permalink'] = rounds2['company_permalink'].str.replace('\W', '')


display(rounds2[rounds2['funding_round_type']=='undisclosed'].shape[0]/rounds2.shape[0] * 100) #percentage of undisclosed rows,only 4.2% hence dropping them
rounds2 = rounds2[rounds2['funding_round_type']!='undisclosed']
display(rounds2[rounds2.duplicated(subset='company_permalink',keep = False)].shape[0])

rounds2 = rounds2[rounds2.raised_amount_usd.notnull()]
display(rounds2)
display(rounds2['company_permalink'].nunique())
```

```python
#Getting the merged dataframes on inner join of the columns.

master_frame1 = pd.merge(companies,rounds2 , how='inner', left_on = companies.permalink,right_on = rounds2.company_permalink)

master_frame1.nunique()

master_frame1 = master_frame1[master_frame1['raised_amount_usd'].between(5000000,15000000,inclusive = True)]

display(master_frame1.nunique())

master_frame1
```

```python
#grouping for funding types.

funding_aggregate = master_frame1.groupby('funding_round_type')
mean = funding_aggregate['raised_amount_usd'].mean()
mean.sort_values(ascending=False)
```

```python
#Visualizing the data

plt.figure(figsize=(14, 8))

# plt.subplot(1, 1, 1)
sns.barplot(x='funding_round_type', y='raised_amount_usd', data=master_frame1,order = ["venture","angel","seed","private_equity"])
plt.show()

# display(master_frame1.head())
```

```python
#Only English speaking countries are present in the CSV
country_code = pd.read_csv("Country_English.csv",encoding = "ISO-8859-1")
country_code = country_code.applymap(lambda s:s.lower() if type(s) == str else s)
country_code = country_code[['Code','country_name']]
# display(country_code)
top9_clean = master_frame1[master_frame1['country_code'].notnull()]

#selecting the funding type
top9_clean = master_frame1[master_frame1['funding_round_type']=='venture']
top9 = pd.merge(top9_clean,country_code,how='inner',left_on=top9_clean.country_code,right_on=country_code.Code)
top9 = top9.drop(['Code'],axis=1)
display(top9.shape[0])
display(top9_clean.shape[0])

# master_frame_aggregate = master_frame1.groupby('country_code')
# master_frame_mean = master_frame_aggregate['raised_amount_usd'].mean()
# display(master_frame_mean.sort_values(ascending=False).head(9))

# top9
top9_sum = top9.groupby('country_name')[['raised_amount_usd']].sum().reset_index()
# top9_sum = top9_aggregate['raised_amount_usd'].sum()
top9_sum = top9_sum.sort_values(by = 'raised_amount_usd',ascending=False).head(9).reset_index()

# top9_sum

# display(top9)
plt.figure(num=None, figsize=(20, 15), dpi=100, facecolor='w', edgecolor='k')
plt.figure(figsize=(10, 10))
sns.barplot(x='raised_amount_usd', y='country_name', data=top9_sum,estimator=sum)
plt.show()
```

```python
#Separating three data frames based on country
D1 = sectors[sectors['country_name']=='united states']
D2 = sectors[sectors['country_name']=='united kingdom']
D3 = sectors[sectors['country_name']=='canada']

#Creating the three groups for count and sum
D1_count = D1.groupby('main_sector')[['raised_amount_usd']].count().reset_index()
D1_count_3 = D1_count.sort_values(['raised_amount_usd'],ascending=False).head(3).reset_index()
D1_sum = D1.groupby('main_sector')[['raised_amount_usd']].sum().reset_index()

D2_count = D2.groupby('main_sector')[['raised_amount_usd']].count().reset_index()
D2_count_3 = D2_count.sort_values(['raised_amount_usd'],ascending=False).head(3).reset_index()
D2_sum = D2.groupby('main_sector')[['raised_amount_usd']].sum().reset_index()

D3_count = D3.groupby('main_sector')[['raised_amount_usd']].count().reset_index()
D3_count_3 = D3_count.sort_values(['raised_amount_usd'],ascending=False).head(3).reset_index()
D3_sum = D3.groupby('main_sector')[['raised_amount_usd']].sum().reset_index()


display(D1_count['raised_amount_usd'].sum(),D2_count['raised_amount_usd'].sum(),D3_count['raised_amount_usd'].sum())
display(D1_sum['raised_amount_usd'].sum(),D2_sum['raised_amount_usd'].sum(),D3_sum['raised_amount_usd'].sum())
display(D1_count.sort_values(['raised_amount_usd'],ascending=False),D2_count.sort_values(['raised_amount_usd'],ascending=False),D3_count.sort_values(['raised_amount_usd'],ascending=False))



D1_company = D1[D1['main_sector']=='others']
D1_company2 = D1[D1['main_sector']=='cleantech / semiconductors']

D2_company = D2[D2['main_sector']=='others']
D2_company2 = D2[D2['main_sector']=='cleantech / semiconductors']

D3_company = D3[D3['main_sector']=='cleantech / semiconductors']
D3_company2 = D3[D3['main_sector']=='others']


#Filtering the results
D1_company = D1_company.groupby('name')[['raised_amount_usd']].sum().reset_index()
display(D1_company.sort_values(['raised_amount_usd'],ascending=False).head(1)) 

D2_company = D2_company.groupby('name')[['raised_amount_usd']].sum().reset_index()
display(D2_company.sort_values(['raised_amount_usd'],ascending=False).head(1))

D3_company = D3_company.groupby('name')[['raised_amount_usd']].sum().reset_index()
display(D3_company.sort_values(['raised_amount_usd'],ascending=False).head(1))



D1_company2 = D1_company2.groupby('name')[['raised_amount_usd']].sum().reset_index()
display(D1_company2.sort_values(['raised_amount_usd'],ascending=False).head(1))

D2_company2 = D2_company2.groupby('name')[['raised_amount_usd']].sum().reset_index()
display(D2_company2.sort_values(['raised_amount_usd'],ascending=False).head(1))

D3_company2 = D3_company2.groupby('name')[['raised_amount_usd']].sum().reset_index()
display(D3_company2.sort_values(['raised_amount_usd'],ascending=False).head(1))

#Dropping the index column as it is not needed.
D1_count_3 = D1_count_3.drop(columns=['index'])
D2_count_3 = D2_count_3.drop(columns=['index'])
D3_count_3 = D3_count_3.drop(columns=['index'])
```

```python
investment_sum = master_frame1['raised_amount_usd'].sum()
master_frame1_gr = master_frame1.groupby('funding_round_type')[['raised_amount_usd']].sum().reset_index()

master_frame1_gr_avg = master_frame1.groupby('funding_round_type')[['raised_amount_usd']].mean().reset_index()

# master_frame1_gr_avg.rename({'raised_amount_type':'funding_average'},axis='columns')
master_frame1_gr_avg.set_axis(['funding_round_type', 'average_fund'], axis=1, inplace=True)


master_frame1_gr['fraction'] = master_frame1_gr['raised_amount_usd']/investment_sum*100

sectors = pd.merge(top9,mapping,how='inner',left_on=top9_category.category_list,right_on=mapping.category_list)

master_frame1_plot1 = pd.merge(master_frame1_gr,master_frame1_gr_avg,how='inner',on = 'funding_round_type')

display(master_frame1_plot1)
```

```python

plt.figure(num=None, figsize=(12, 8), dpi=100, facecolor='w', edgecolor='k')
sns.barplot(x='funding_round_type', y='average_fund', hue="fraction", data=master_frame1_plot1, estimator=np.mean,order=["venture","private_equity","angel","seed"])
plt.show()
```
